```python
# get some properties from user
class course:
    def __init__(self, name):
        self.__name = name # this makes this property private
        self.students = []
    
    # add some more functionality
    def add_students(self, student):
        self.students.append(student)
        self.__write_student(student)
        
    def student_count(self):
        return len(self.students)
    
    def __write_student(self, student):
        print(f'Hello {student}!')
```

```python
c1 = course('Python')
```

```python
c1.add_students('Joe')
```

```python
c1.student_count()
```

```python
c1.add_students('Matt')
```

```python
c1.student_count()
```

## Magic 8 Ball Class

```python
import random, csv

class magic8ball:
    # do I need to add any properties? Should they be private or public?
    
    # do I need to add any functions or methods?
    
    # add the init function
    def __init__(self, name):
        self.__name = name
        self.__mQuestions = []
        self.__response_list = [
            'Yes, absolutely',
            'Not a chance',
            'Get a life',
            'No doubt',
            'Si Dios quiere',
            'Only if you really want it',
            'How much are you willing to pay me for it?',
            'That would be awesome, wouldnt it?',
            'Yes, that will happen tomorrow',
            'You were destined for it'
        ]
        self.__welcome(self.__name)
        self.__start_game()
        
    
    # create a private method for the game
    def __start_game(self):
        question = 'Starter'
        while question != '':
            question = input('Please enter your question: ')
            if question != '':
                self.__mQuestions.append(question)
                n = random.randint(0,9)
                print(self.__response_list[n])
            else:
                print('Thank you for playing!')
                self.__write_questions()
            
    # create private method to welcome user
    def __welcome(self, name):
        print(f'Welcome, {name}!! Thanks for playing. Press enter when done.')
        
    # create a private method to write to a csv file
    def __write_questions(self):
        f = open('magic_questions.csv', 'a')
        wrt = csv.writer(f)
        for q in self.__mQuestions:
            wrt.writerow([q])
        f.close()

        
```
